$\large{\textbf{1. Introduction}}$

This tutorial discusses the basic concepts of neural networks (NNs) or artificial neural networks (ANNs) and illustrates different activation functions in NNs using numerical examples. NN or ANN is a network of artificial neurons that are arranged in layers. NN can be considered as a nonlinear function from a mathematical point of view, which can be used for obtaining nonlinear models from data. And the activation functions are the nonlinear component in NN. NN is a popular topic in artificial intelligence (AI) and machine learning (ML) because of its ability to model complex relationships in data. In this tutorial, we discuss the basic structure of NN and illustrate how we can use NN for approximating arbitrary functions. Before moving on to NN, let's give a brief introduction to AI and ML, which is coming up next.

${\textbf{1.2. From ML to NN}}$

Machine learning (ML)  deals with solving practical problems by constructing or learning models using data.  The model can be a function or an algorithm that transforms the data into a suitable form for interpretation.  The term machine learning indicates that here the machine is programmed to learn a model using the data or available information. And by machine, here we mean a computer.

Based on the nature of the available data set and objective of the problem AI and ML algorithms can be grouped into the following categories.
1. $\color{red}{\textbf{Supervised learning}}$: used for obtaining  model of a set of data containing the inputs and outputs ($\textbf{labels}$). This can be further divided as
      * $\textit{Regression}$: in which outputs or labels are continuous variables.
      * $\textit{Classification}$: in which the outputs or labels are from a discrete set with finite elements.
2. $\color{red}{\textbf{Unsupervised learning}}$: used for obtaining models that extract useful information
from the input ($\textbf{unlabelled data}$). This can be further divided as
      * $\textit{Dimensionality reduction}:$ which deals with reducing the dimension or size of the data while preserving the relevant information.
      * $\textit{Clustering}:$ which is associated with learning categories or patterns and
      categorizing the input data into different clusters.
3. $\color{red}{\textbf{Reinforcement learning}}$: is used to learn a sequence of decisions or a  policy using data. This can be further divided as
      * $\textit{Model-based algorithms}:$ which uses the model information for computing the policy.
      * $\textit{Model free algorithms}:$ which does not require any model information.

NN can be used for constructing models that can solve various problems in the above-mentioned categories: regression, classification, dimensionality reduction, clustering, etc.
In this tutorial, we will be focusing on regression problems.  In regression problems, the labels are from a continuous set.
The objective is to find a model that map the input feature vector to a continuous label:

\begin{equation}
    \hat{y}=f(x,P) \hspace{17.5cm} (1)
\end{equation}

where $x \in \mathbb{R}^{n}$ is the input vector or observation vector, $\hat{y} \in \mathbb{R}^{m}$ is the predicted output, $f:\mathbb{R}^{n}\rightarrow \mathbb{R}^{m}$ is the model, and $P$ is a vector or matrix containing the free parameters in the model. In regression problems, our objective is to train or learn a model $f$ that fits with the available input-output data. Here by training a model, we mean finding the best choice of free parameters, i.e. the optimal $P.$
The model is then used for predicting the output for a new input vector. In general, based on the model, regression problems can be further divided into two categories:

  1. $\color{red}{\textbf{Linear}}$: in which the function $f$ is linear.
  2. $\color{red}{\textbf{Nonlinear}}$: in which the function $f$ is nonlinear.

In linear regression, we fit the input-output data with a linear function or model. This means a line in the case of univariate input-output data and a hyperplane in the case of multivariate regression. In general, a linear input-output model can be represented as
\begin{equation}
     \hat{y}=f_{L}(x,P)=Ax+b  \hspace{15cm} (2)
\end{equation}
where $A\in \mathbb{R}^{m \times n}$ is the weight and $b\in \mathbb{R}^{m}$ is the bias which are the parameters in the model. Note that in the case of univariate regression the weight and bias terms become scalars, which correspond to the slope and intercept of the line.

Now in linear regression, our objective is to find the optimal value of the weight $A$ and bias $b$ that fits the data in the best way. For that, we can use the available data or training data, which consists of:


 1.   $\color{red}{\textbf{Inputs}}$: consists of $N$ observations $\{ x_{1}, x_{2},...,x_{N} \}$.
 2.   $\color{red}{\textbf{Outputs}}$: consists of $N$ labels $\{y_{1}, y_{2}, ..., y_{N} \}$.


To illustrate linear regression, let's consider a univariate example in which the input-output data is related by a linear equation:
\begin{equation}
    y=3x+2
\end{equation}
which we consider as the actual model. Here by actual model, we mean the true relationship between the input and output, which is normally unknown. Therefore, our objective is to find a model that approximates the actual model and can be used for predicting the outputs. For that, we use the available information or the training data which consists of the training inputs and outputs.
In this example, the training data is generated by adding some randomness to the actual model. We consider the
training input $\{ x_{1},...,x_{100} \}$ containing 100 points in the range $[0, 10]$.  The training output is generated as
\begin{equation}
y_{tr}=3x_{tr}+2 + r
\end{equation}
where $r$ is a Gaussian random variable with mean zero and variance one.  Note that in the case of univariate regression, the weight and bias terms become scalars, which correspond to the slope and intercept of the line. In this case, the data follows a linear trend, hence, a linear model is sufficient. However, in most of the practical problems, the data follows a nonlinear trend. In that case, a nonlinear model is required for an accurate representation of the data. This is illustrated next with two examples in which the data is generated using the following nonlinear relationships:
\begin{equation}
    \begin{aligned}
        &y=e^{0.5x}\\
        &y=sin(x)
    \end{aligned}
\end{equation}
In this case, the predicted output does not match the actual output. This is because the predicted outputs follow a line (since we used the linear regression model), whereas the actual outputs follow a nonlinear trend. Consequently, here a linear model is not sufficient and we need a nonlinear model for more accurate predictions.
The illustration of these examples in Python is given below.

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
A=3; b=2
X = np.linspace(0, 10, 100)       # X contains 100 equally spaced values between 0 and 10
Y = A * X + b
rng = np.random.RandomState(1)
Xtr = np.linspace(0, 10, 20)
Ytr = A * Xtr + b + rng.randn(len(Xtr))
# create a linear regression model
f_L = LinearRegression()             #f_L denotes linear model
f_L.fit(Xtr[:, np.newaxis], Ytr)
YhatL = f_L.predict(X[:, np.newaxis])  # YhatL denotes predicted output

# Example of data with nonlinear trend 1
rng = np.random.RandomState(1)
Xa = np.linspace(0, 10, 100)
Ya = np.exp(0.5*X)
rng = np.random.RandomState(1)
Xtra = np.linspace(0, 10, 20)
Ytra = np.exp(0.5*Xtra) + 5*rng.randn(len(Xtra))
f_La = LinearRegression()             #f_L denotes linear model
f_La.fit(Xtra[:, np.newaxis], Ytra)
YhatLa = f_La.predict(Xa[:, np.newaxis])  # YhatL denotes predicted output

# Example of data with nonlinear trend 1
Xb = np.linspace(0, 10, 100)
Yb = np.sin(X)
Xtrb = np.linspace(0, 10, 20)
Ytrb = np.sin(Xtrb) + 0.2*rng.randn(len(Xtrb))
f_Lb = LinearRegression()
f_Lb.fit(Xtrb[:, np.newaxis], Ytrb)
YhatLb = f_Lb.predict(Xb[:, np.newaxis])

# plotting results
fig, axs = plt.subplots(1, 4, figsize =(16, 3))
axs[0].scatter(Xtr, Ytr, c = 'r');
axs[0].plot(X, Y, c = 'black')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend(['$y_{tr}$','$y$'])
axs[1].scatter(Xtr, Ytr, c = 'red')
axs[1].plot(X, Y, c = 'black')
axs[1].plot(X, YhatL, c = 'green')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend(['$y_{tr}$','$y$', '$\hat{y}_{L}$'])
axs[2].scatter(Xtra, Ytra, c = 'red')
axs[2].plot(Xa, Ya, c = 'black')
axs[2].plot(Xa, YhatLa, c = 'green')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].legend(['$y_{tr}$','$y$', '$\hat{y}_{L}$'])
axs[3].scatter(Xtrb, Ytrb, c = 'red')
axs[3].plot(Xb, Yb, c = 'black')
axs[3].plot(Xb, YhatLb, c = 'green')
axs[3].set_xlabel('x')
axs[3].set_ylabel('y')
axs[3].legend(['$y_{tr}$','$y$', '$\hat{y}_{L}$'])
plt.savefig('fig15.svg')
```

In the above figure,  the red dots show the training output, the black line shows the actual output and the green line shows the predicted output. In the second subplot, we can see that the predicted output closely matches with the actual output. In this case, the data follows a linear trend, hence, a linear model is sufficient. However, in most of the practical problems, the data follows a nonlinear trend. In that case, a nonlinear model is required for an accurate representation of the data. This situation is illustrated in subplots 3 and 4. In such cases, NNs can be used for constructing sufficiently accurate nonlinear models and are discussed next.




$\large{\textbf{2. Neural Network (NN)}}$  

 NN is a network of neurons that are arranged in layers.  Each layer transforms its input data and gives an output. And the output of the layer will be the input of the next layer.
Therefore from a mathematical point of view, NN is a $\textbf{composite function}$ of the following form:

\begin{equation}
    \hat{y}=f_{L+1}(f_{L}(...(f_{1}(f_{0}(x)))) \hspace{11.75cm} (3)
\end{equation}

where $f_{0},f_{1},f_{2},...,f_{L+1}$ are $\textbf{layers}$ of the NN which are composite functions of the form:

\begin{equation}
    f_{i}(x)=\sigma(Ax+b)  \hspace{14cm} (4)
\end{equation}

The functions $f_{i}$ consist of two components:

  1. $\color{red}{\textbf{Affine function}}$: is the linear component which is of the form $Ax+b.$

  2.   $\color{red}{\textbf{Activation function}}$: is the nonlinear component. One of the popular activation function is the ReLU
      (Rectified Linear Unit) function which is defined as below:
        \begin{equation}
        \sigma(x)=\max(0,x)= \begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}  \hspace{8.75cm} (5)
        \end{equation}

${\textbf{2.1. Neuron}}$

 Neurons that we are considering here are artificial neurons which are inspired by biological neurons. The biological neurons are signal processing units and are the basic building blocks of the biological neural network (BNN) or nervous system in the brain. Similarly, artificial neurons are data processing units and are the basic building blocks of the NN or ANN. Artificial neurons can be mathematically represented as
    \begin{equation}
    y=\sigma(Ax+b)
\end{equation}
The input to the neuron or $x$ can be a scalar or a vector. The output to the neuron or $y$ will be a scalar.


${\textbf{2.2. Layer}}$

$\color{red}{\textbf{Layer}}$ is a collection of neurons that takes the same inputs. In general, layers can be considered as vector-valued functions and neurons can be considered as scalar-valued functions.  Layers can be grouped into 3 categories:


   1.  $\color{red}{\textbf{Input layer}}$: in which each neuron has only one input and it simply passes the input. i.e. $A=1,$ $b=0,$ and $\sigma(Ax+b)=x.$
   2. $\color{red}{\textbf{Hidden layer}}$: in which each neuron can have more than one input and gives a scalar output. The activation function is normally noninear: ReLU, tanh, etc.
   2.  $\color{red}{\textbf{Output layer}}$: in which each neuron can have more than one input and  gives a scalar output. The activation function is normally linear.
  

$\large{\textbf{3. Activation functions}}$

In this section, we explore other activation functions that are used in NNs. The activation functions that are commonly used in NNs are listed below:
 1.  $\color{red}{\textbf{Linear}}$ function:
      \begin{equation}
        \sigma(x)=ax
      \end{equation}
 2. $\color{red}{\textbf{ReLU}}$ (Rectified Linear Unit) function:
    \begin{equation}
        \sigma(x)=\max(0,x)= \begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}
 3. $\color{red}{\textbf{Leaky ReLU}}$ function:
    \begin{equation}
        \sigma(x)= \begin{cases}
        ax, \hspace{0.3cm} a>0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}   
 4. $\color{red}{\textbf{Modulus}}$ function:
    \begin{equation}
        \sigma(x)=|x|= \begin{cases}
        -x, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
    \end{equation}   

 5. $\color{red}{\textbf{Step}}$ function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        0, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
     \end{equation}
 6. $\color{red}{\textbf{Signum}}$ function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        -1, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x >0
        \end{cases}
     \end{equation}    

 7. $\color{red}{\textbf{Saturation}}$ (Satlin) function:
     \begin{equation}
        \sigma(x)=\begin{cases}
        -1, \hspace{0.3cm} & if \hspace{0.1cm} x\leq 0\\
        x, \hspace{0.3cm} & if \hspace{0.1cm} -1< x < 1\\
        1, \hspace{0.3cm} &if \hspace{0.1cm} x \geq 1
        \end{cases}
     \end{equation}

 8. $\color{red}{\textbf{Sigmoid}}$ function:
    \begin{equation}
      \sigma(x)=\frac{1}{1+e^{-x}}
    \end{equation}
 9.   $\color{red}{\textbf{Swish}}$  function:
    \begin{equation}
        \sigma(x)= x\hspace{.1cm} sigmoid(\beta x)=\frac{x}{1+e^{-\beta x}}
    \end{equation}

 10.   $\color{red}{\textbf{Tanh}}$ (hyperbolic tan) function:
    \begin{equation}
        \sigma(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}
    \end{equation}



The graph of different activation functions are plotted next.

```python
m1=100
beta=0.5
X = np.linspace(-10, 10, 100)
Ya=np.zeros(len(X))
Yb=np.zeros(len(X))
Yc=np.zeros(len(X))
Yd=np.zeros(len(X))
Ye=np.zeros(len(X))
Yf=np.zeros(len(X))
Yg=np.zeros(len(X))
Yh=np.zeros(len(X))
for i in range(len(X)):
  Ya[i] = np.max([X[i],0])                             # ReLU
  Yb[i] = np.abs(X[i])                                 # Modulus
  Yc[i] = np.heaviside(X[i],0)                         # Step
  Yd[i] = np.sign(X[i])                                # Signum
  Ye[i] = np.clip(X[i],-1,1)                           # Saturation
  Yf[i] = np.divide(1, 1+ np.exp(-X[i]))               # Sigmoid
  Yg[i] = np.divide(X[i], 1+ np.exp(-beta*X[i]))       # Swish
  Yh[i] = np.tanh(X[i])                                # Tanh
fig, axs = plt.subplots(2, 4, figsize =(16, 6))
axs[0, 0].plot(X, Ya, c = 'r');
axs[0, 0].set_title("ReLU")
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('$\sigma(x)$')
axs[0, 1].plot(X, Yb, c = 'r');
axs[0, 1].set_title("Modulus")
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('$\sigma(x)$')
axs[0, 2].plot(X, Yc, c = 'r');
axs[0, 2].set_title("Step")
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('$\sigma(x)$')
axs[0, 3].plot(X, Yd, c = 'r');
axs[0, 3].set_title("Signum")
axs[0, 3].set_xlabel('x')
axs[0, 3].set_ylabel('$\sigma(x)$')
axs[1, 0].plot(X, Ye, c = 'r');
axs[1, 0].set_title("Saturation")
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('$\sigma(x)$')
axs[1, 1].plot(X, Yf, c = 'r');
axs[1, 1].set_title("Sigmoid")
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('$\sigma(x)$')
axs[1, 2].plot(X, Yg, c = 'r');
axs[1, 2].set_title("Swish")
axs[1, 2].set_xlabel('x')
axs[1, 2].set_ylabel('$\sigma(x)$')
axs[1, 3].plot(X, Yh, c = 'r');
axs[1, 3].set_title("Tanh")
axs[1, 3].set_xlabel('x')
axs[1, 3].set_ylabel('$\sigma(x)$')
plt.savefig('fig7.svg')
```

$\textbf{3.2. NN with more than one neuron per layer}$

Here we are considering NNs with more than one neuron per layer. To simplify the illustration we consider examples with only one hidden layer as shown in the figure below.

Here we denote the output of the first neuron in first hidden layer as $y_{11}$ and the second neuron as $y_{12}$ and so on. Let's start with the following NNs with ReLU activation functions:
\begin{equation}
    \begin{aligned}
      &(a) \hspace{.5cm} y=\text{ReLU}(2 x+ 1)\\
        &(b) \hspace{.5cm}y=\text{ReLU}(- x- 2)\\
        &(c) \hspace{.5cm}y=\text{ReLU}(2 x+ 1)+\text{ReLU}(- x - 1)\\
        &(d) \hspace{.5cm}y=\text{ReLU}(2 x+ 1)+\text{ReLU}(- x- 2)+\text{ReLU}(4 x- 20)
    \end{aligned}
\end{equation}
which corresponds to NNs with one neuron (a,b), two neurons (c), and three neurons (d).

```python
X = np.linspace(-10, 10, 100)
Y11=np.zeros(len(X))
Y12=np.zeros(len(X))
Y13=np.zeros(len(X))
Ya=np.zeros(len(X))
Yb=np.zeros(len(X))
for i in range(len(X)):
 Y11[i] = np.max([(2*X[i]+1),0])
 Y12[i] = np.max([(-1*X[i]-2),0])
 Y13[i] = np.max([(4*X[i]-20),0])
 Ya[i]  = Y11[i]+Y12[i]
 Yb[i]  = Y11[i]+Y12[i]+Y13[i]
fig, axs = plt.subplots(1, 4, figsize =(16, 3))
axs[0].plot(X, Y11, c = 'r');
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[1].plot(X, Y12, c = 'r')
axs[1].set_xlabel('x')
axs[1].set_ylabel('$y$')
axs[2].plot(X, Ya, c = 'r')
axs[2].set_xlabel('x')
axs[2].set_ylabel('$y$')
axs[3].plot(X, Yb, c = 'r')
axs[3].set_xlabel('x')
axs[3].set_ylabel('$y$')
plt.savefig('fig7.svg')
```

Here with two neurons, we obtained a folded line with two folds. Further with three neurons, we obtained a folded line with three folds. By adding more neurons, we can add more folds in the line. Folded lines are also called a piecewise linear curve. Therefore with a ReLU activation function, we can obtain piecewise linear approximations of continuous functions. Moreover, we can also use other activation functions like sigmoid, tanh, etc for getting smooth approximations. This will be illustrated in the next section.   

```python
m1=3
#m1=100
X = np.linspace(-10, 10, 100)
Y1a=np.zeros(m1)
Y1b=np.zeros(m1)
Y1c=np.zeros(m1)
Y1d=np.zeros(m1)
Y1e=np.zeros(m1)
Y1f=np.zeros(m1)
Y1g=np.zeros(m1)
Y1h=np.zeros(m1)
Ya=np.zeros(len(X))
Yb=np.zeros(len(X))
Yc=np.zeros(len(X))
Yd=np.zeros(len(X))
Ye=np.zeros(len(X))
Yf=np.zeros(len(X))
Yg=np.zeros(len(X))
Yh=np.zeros(len(X))
rng = np.random.RandomState(5)    # By changing the RandomState input argument we can get different outputs
A1=rng.randn(m1)
b1=rng.randn(m1)
A2=rng.randn(m1)
b2=rng.randn(1)
for i in range(len(X)):
 for j in range(m1):
  Y1a[j] = np.max([(A1[j]*X[i]+b1[j]),0])                                        # ReLU
  Y1b[j] = np.abs(A1[j]*X[i]+b1[j])                                              # Modulus
  Y1c[j] = np.heaviside(A1[j]*X[i]+b1[j],0)                                      # Step
  Y1d[j] = np.sign(A1[j]*X[i]+b1[j])                                             # Signum
  Y1e[j] = np.clip(A1[j]*X[i]+b1[j],-1,1)                                        # Saturation
  Y1f[j] = np.divide(1, 1+ np.exp(-(A1[j]*X[i]+b1[j])))                          # Sigmoid
  Y1g[j] = np.divide((A1[j]*X[i]+b1[j]), 1+ np.exp(-(beta*(A1[j]*X[i]+b1[j]))))  # Swish
  Y1h[j] = np.tanh(A1[j]*X[i]+b1[j])                                             # Tanh
 Ya[i]  = np.sum(np.multiply(A2,Y1a)+b2)
 Yb[i]  = np.sum(np.multiply(A2,Y1b)+b2)
 Yc[i]  = np.sum(np.multiply(A2,Y1c)+b2)
 Yd[i]  = np.sum(np.multiply(A2,Y1d)+b2)
 Ye[i]  = np.sum(np.multiply(A2,Y1e)+b2)
 Yf[i]  = np.sum(np.multiply(A2,Y1f)+b2)
 Yg[i]  = np.sum(np.multiply(A2,Y1g)+b2)
 Yh[i]  = np.sum(np.multiply(A2,Y1h)+b2)

fig, axs = plt.subplots(2, 4, figsize =(16, 6))
axs[0, 0].plot(X, Ya, c = 'r');
axs[0, 0].set_title("ReLU")
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 1].plot(X, Yb, c = 'r');
axs[0, 1].set_title("Modulus")
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 2].plot(X, Yc, c = 'r');
axs[0, 2].set_title("Step")
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('y')
axs[0, 3].plot(X, Yd, c = 'r');
axs[0, 3].set_title("Signum")
axs[0, 3].set_xlabel('x')
axs[0, 3].set_ylabel('y')
axs[1, 0].plot(X, Ye, c = 'r');
axs[1, 0].set_title("Saturation")
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 1].plot(X, Yf, c = 'r');
axs[1, 1].set_title("Sigmoid")
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 2].plot(X, Yg, c = 'r');
axs[1, 2].set_title("Swish")
axs[1, 2].set_xlabel('x')
axs[1, 2].set_ylabel('y')
axs[1, 3].plot(X, Yh, c = 'r');
axs[1, 3].set_title("Tanh")
axs[1, 3].set_xlabel('x')
axs[1, 3].set_ylabel('y')
```

$\large{\textbf{4. Neural Network Regression}}$

In Neural Network regression, the model is an NN:
\begin{equation}
    \hat{y}=f_{NN}(x,P)=f_{L+1}(f_{L}(...(f_{0}(x))))
\end{equation}
where $f_{0},f_{1},...,f_{L+1}$ are layers of the NN. Here the parameter P is the weights and biases of each layer  in the NN which are to be optimized. And for that we use some optimization algorithms which uses the training data and computes the optimal parameters which minimizes the cost function $J(y_{tr}-\hat{y})$.  

Here we illustrate the neural network regression using two examples in which the data is generated from exponential and sinusoidal functions.
The training input  $\{x_{1},...,x_{20} \}$ consists of 20 data points spaced uniformly in range $[0, 10]$.
The training output for the examples are generated as
\begin{equation}
y_{tr}=e^{0.5x_{tr}}+5 r
\end{equation}
\begin{equation}
y_{tr}=sin(x_{tr})+0.2 r
\end{equation}
where $r$ is a Gaussian random variable with mean 0 and variance 1.

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
rng = np.random.RandomState(1)
Xa = np.linspace(0, 10, 100)
Ya = np.exp(0.5*Xa)
Xtra = np.linspace(0, 10, 20)
Ytra = np.exp(0.5*Xtra) + 5*rng.randn(len(Xtra))
f_NNa = MLPRegressor(hidden_layer_sizes=(5,5,5),activation="relu" , solver='lbfgs', max_iter=10000)
#lbfgs:  Limited-memory Broyden-Fletcher-Goldfarb-Shanno    # for displaying error use verbose=True
f_NNa.fit(Xtra[:, np.newaxis], Ytra)
YhatNNa = f_NNa.predict(Xa[:, np.newaxis])

Xb = np.linspace(0, 10, 100)
Yb = np.exp(0.5*Xb)
Xtrb = np.linspace(0, 10, 20)
Ytrb = np.exp(0.5*Xtrb) + 5*rng.randn(len(Xtrb))
f_NNb = MLPRegressor(hidden_layer_sizes=(5,5,5),activation="tanh" , solver='lbfgs', max_iter=10000)
f_NNb.fit(Xtrb[:, np.newaxis], Ytrb)
YhatNNb = f_NNb.predict(Xb[:, np.newaxis])

Xc = np.linspace(0, 10, 100)
Yc = np.sin(Xc)
Xtrc = np.linspace(0, 10, 20)
Ytrc = np.sin(Xtrc) + 0.2*rng.randn(len(Xtrc))
f_NNc = MLPRegressor(hidden_layer_sizes=(5,5,5),activation="relu" , solver='lbfgs', max_iter=10000)
f_NNc.fit(Xtrc[:, np.newaxis], Ytrc)
YhatNNc = f_NNc.predict(Xc[:, np.newaxis])

Xd = np.linspace(0, 10, 100)
Yd = np.sin(Xd)
Xtrd = np.linspace(0, 10, 20)
Ytrd = np.sin(Xtrd) + 0.2*rng.randn(len(Xtrd))
f_NNd = MLPRegressor(hidden_layer_sizes=(5,5,5),activation="tanh" , solver='lbfgs', max_iter=10000)
f_NNd.fit(Xtrd[:, np.newaxis], Ytrd)
YhatNNd = f_NNd.predict(Xd[:, np.newaxis])

# plotting results
fig, axs = plt.subplots(1, 4, figsize =(16, 3))
axs[0].scatter(Xtra, Ytra, c = 'red')
axs[0].plot(Xa, Ya, c = 'black')
axs[0].plot(Xa, YhatNNa, c = 'green')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend(['$y_{tr}$','$y$', '$\hat{y}_{NN}$'])
axs[1].scatter(Xtrb, Ytrb, c = 'red')
axs[1].plot(Xb, Yb, c = 'black')
axs[1].plot(Xb, YhatNNb, c = 'green')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend(['$y_{tr}$','$y$', '$\hat{y}_{NN}$'])
axs[2].scatter(Xtrc, Ytrc, c = 'red')
axs[2].plot(Xc, Yc, c = 'black')
axs[2].plot(Xc, YhatNNc, c = 'green')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].legend(['$y_{tr}$','$y$', '$\hat{y}_{NN}$'])
axs[3].scatter(Xtrd, Ytrd, c = 'red')
axs[3].plot(Xd, Yd, c = 'black')
axs[3].plot(Xd, YhatNNd, c = 'green')
axs[3].set_xlabel('x')
axs[3].set_ylabel('y')
axs[3].legend(['$y_{tr}$','$y$', '$\hat{y}_{NN}$'])
plt.savefig('fig15.svg')
```

In the above figure, the red dots show the training output, the black curve shows the actual output and the green curve shows the predicted output. Here we can see that the predicted output closely matches with the actual output. Further, the predicted output curve is piecewise linear in the case of ReLU activation function. Whereas for tanh activation function, the predicted output is a smooth curve as expected.
